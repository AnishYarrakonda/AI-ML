```python
# 0. Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)
```

```python
# 1. Minimal Plot Skeleton
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
# 2. Line Plots (single + multiple + legend)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label="sin")
ax.plot(x, y2, label="cos")
ax.set_title("Line Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

```

```python
# 3. Scatter Plots (color, size, alpha)
x = rng.normal(size=200)
y = rng.normal(size=200)
colors = np.hypot(x, y)
sizes = (np.abs(x) + 0.1) * 80

fig, ax = plt.subplots()
ax.scatter(x, y, c=colors, s=sizes, alpha=0.6)
ax.set_title("Scatter Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
# 4. Bar Charts (vertical, horizontal, categorical)
labels = ["A", "B", "C", "D"]
values = [5, 2, 7, 4]

fig, ax = plt.subplots()
ax.bar(labels, values)
ax.set_title("Bar Chart")
ax.set_xlabel("category")
ax.set_ylabel("value")
plt.show()

fig, ax = plt.subplots()
ax.barh(labels, values)
ax.set_title("Horizontal Bar Chart")
ax.set_xlabel("value")
ax.set_ylabel("category")
plt.show()
```

```python
# 5. Histograms (bins, multiple, density)
data1 = rng.normal(loc=0, scale=1, size=1000)
data2 = rng.normal(loc=2, scale=1.5, size=1000)

fig, ax = plt.subplots()
ax.hist(data1, bins=30, alpha=0.7, label="A")
ax.hist(data2, bins=30, alpha=0.7, label="B")
ax.set_title("Histograms")
ax.set_xlabel("value")
ax.set_ylabel("count")
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.hist(data1, bins=30, density=True, alpha=0.7)
ax.set_title("Normalized Histogram")
ax.set_xlabel("value")
ax.set_ylabel("density")
plt.show()
```

```python
# 6. Titles, Labels, Legends (required)
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [2, 1, 3], label="line")
ax.set_title("Title")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# 7. Figure Size & Saving
x = np.linspace(0, 4, 200)
y = np.exp(-x)

fig, ax = plt.subplots(figsize=(6, 3), dpi=120)
ax.plot(x, y, label="decay")
ax.set_title("Sized Figure")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.tight_layout()
plt.savefig("matplotlib_sized_plot.png")
plt.show()
```

```python
# 8. Subplots (grids, shared axes)
fig, axes = plt.subplots(2, 2, sharex=True, figsize=(6, 5))

x = np.linspace(0, 2*np.pi, 200)

axes[0, 0].plot(x, np.sin(x), label="sin")
axes[0, 1].plot(x, np.cos(x), label="cos")
axes[1, 0].plot(x, np.tan(x), label="tan")
axes[1, 1].plot(x, np.sin(2*x), label="sin2x")

for ax in axes.ravel():
    ax.set_title("subplot")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 9. Axis Limits & Scales
x = np.logspace(0, 3, 100)
y = np.log10(x)

fig, ax = plt.subplots()
ax.plot(x, y, label="log10")
ax.set_title("Log Scale X")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_xscale('log')
ax.legend()
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y, label="linear")
ax.set_title("Linear Scale")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# 10. Grid & Ticks
x = np.linspace(0, 10, 50)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label="sin")
ax.set_title("Grid & Ticks")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_xticklabels(["0", "2", "4", "6", "8", "10"], rotation=30)
ax.legend()
plt.show()
```

```python
# 11. Annotations & Reference Lines
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, label="sin")
ax.axhline(0, color="gray", linestyle="--")
ax.axvline(np.pi, color="red", linestyle=":")
ax.annotate("pi", xy=(np.pi, 0), xytext=(np.pi+1, 0.5),
            arrowprops=dict(arrowstyle="->"))
ax.set_title("Annotations")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# 12. Plotting NumPy & Pandas Directly
x = np.arange(5)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, label="np array")
ax.set_title("NumPy Arrays")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()

s = pd.Series([3, 1, 4, 1, 5], name="series")
fig, ax = plt.subplots()
s.plot(ax=ax, title="Pandas Series")
ax.set_xlabel("index")
ax.set_ylabel("value")
plt.show()

 df = pd.DataFrame({"a": [1, 2, 3], "b": [3, 2, 1]})
fig, ax = plt.subplots()
df.plot(ax=ax, title="Pandas DataFrame")
ax.set_xlabel("index")
ax.set_ylabel("value")
plt.show()
```

```python
# 13. Common Mistakes (shown via code)
# Overwriting plots: reusing the same axes without clearing
x = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label="sin")
ax.plot(x, np.cos(x), label="cos")
ax.set_title("Two Lines on Same Axes")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()

# Forgetting ax (stateful interface)
plt.plot(x, np.sin(x))
plt.title("Stateful Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 14. Object-Oriented Control
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y, label="sin")
ax.set_title("OO Control")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()

# Update data
line.set_ydata(np.cos(x))
ax.set_title("Updated to Cos")
plt.show()
```

```python
# 15. Advanced Subplots (gridspec, insets, twinx)
from matplotlib.gridspec import GridSpec
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig = plt.figure(figsize=(6, 4))
gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])

x = np.linspace(0, 2*np.pi, 200)
ax1.plot(x, np.sin(x), label="sin")
ax1.set_title("Gridspec Top")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

ax2.plot(x, np.cos(x), label="cos")
ax2.set_title("Bottom Left")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.legend()

ax3.plot(x, np.tan(x), label="tan")
ax3.set_title("Bottom Right")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.legend()

axins = inset_axes(ax1, width="30%", height="30%", loc="upper right")
axins.plot(x, np.sin(2*x))
axins.set_title("Inset")

ax_twin = ax1.twinx()
ax_twin.plot(x, 0.5*np.cos(x), color="orange", label="cos (scaled)")
ax_twin.set_ylabel("twin y")

plt.tight_layout()
plt.show()
```

```python
# 16. Colormaps
X = np.random.randn(50, 50)

fig, ax = plt.subplots()
im = ax.imshow(X, cmap="viridis")
ax.set_title("Continuous Colormap")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.colorbar(im, ax=ax)
plt.show()
```

```python
# 17. Heatmaps / Images (imshow)
M = np.arange(100).reshape(10, 10)

fig, ax = plt.subplots()
im = ax.imshow(M, cmap="magma", aspect="auto", vmin=0, vmax=100)
ax.set_title("Heatmap")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.colorbar(im, ax=ax)
plt.show()
```

```python
# 18. Style Control
plt.style.use("seaborn-v0_8")

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x), linestyle="--", marker="o", label="sin")
ax.plot(x, np.cos(x), linestyle=":", marker="s", label="cos")
ax.set_title("Styles")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()

# Reset to default
plt.style.use("default")
```

```python
# 19. Layout Management
fig, axes = plt.subplots(2, 2, figsize=(6, 4), constrained_layout=True)

x = np.linspace(0, 2*np.pi, 200)
for ax in axes.ravel():
    ax.plot(x, np.sin(x), label="sin")
    ax.set_title("subplot")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.show()

fig, axes = plt.subplots(2, 2, figsize=(6, 4))
for ax in axes.ravel():
    ax.plot(x, np.cos(x), label="cos")
    ax.set_title("subplot")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# 20. Performance Demo
N = 1_000_000
x = np.linspace(0, 100, N)
y = np.sin(x)

# Downsampling
x_ds = x[::1000]
y_ds = y[::1000]

fig, ax = plt.subplots()
ax.plot(x_ds, y_ds, label="downsampled")
ax.set_title("Downsampled Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()

import timeit

time_full = timeit.timeit(lambda: plt.plot(x, y), number=1)
plt.clf()

time_ds = timeit.timeit(lambda: plt.plot(x_ds, y_ds), number=1)
plt.clf()

time_full, time_ds
```
